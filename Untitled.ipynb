```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# ---------------- DATA COLLECTION ----------------
```

```python
wine_dataset = pd.read_csv('WineQT.csv')

print(wine_dataset.shape)
print(wine_dataset.head())
print(wine_dataset.isnull().sum())
```

```python
# ---------------- DATA ANALYSIS ----------------
```

```python
print(wine_dataset.describe())

sns.catplot(x='quality', data=wine_dataset, kind='count')
plt.show()

plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='volatile acidity', data=wine_dataset)
plt.show()

plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='citric acid', data=wine_dataset)
plt.show()
```

```python
# ---------------- CORRELATION ----------------
```

```python
correlation = wine_dataset.corr(numeric_only=True)

plt.figure(figsize=(10,10))
sns.heatmap(correlation, annot=True, cmap='Blues', fmt='.1f')
plt.show()
```

```python
# ---------------- DATA PROCESSING ----------------
```

```python
X = wine_dataset.drop('quality', axis=1)
Y = wine_dataset['quality'].apply(lambda y: 1 if y >= 7 else 0)
```

```python
# ---------------- TRAIN TEST SPLIT ----------------
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=3
)
```

```python
# ---------------- MODEL TRAINING ----------------
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, Y_train)
```

```python
# ---------------- MODEL EVALUATION ----------------
```

```python
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy:", accuracy)
```

```python
# ---------------- PREDICTION SYSTEM ----------------
```

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5,11.5)

input_df = pd.DataFrame([input_data], columns=X.columns)

prediction = model.predict(input_df)

if prediction[0] == 1:
    print("Good Quality Wine")
else:
    print("Bad Quality Wine")
```
